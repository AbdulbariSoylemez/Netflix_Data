```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
veri=pd.read_csv("netflix_titles.csv")
veri.head()
```

```python
veri.shape
```

```python
veri.columns
```

```python
veri.isnull().sum()
```

```python
veri.nunique()
```

```python
data=veri.copy()
data.head()
```

```python
data.shape
```

```python
data=data.dropna()
data.shape
```

```python
sns.countplot(veri["type"])
fig=plt.gcf()
fig.set_size_inches(5,5)
plt.title("tür")
```

```python
sns.countplot(veri["rating"])
sns.countplot(veri["rating"]).set_xticklabels(sns.countplot(veri["rating"]).get_xticklabels(),rotation=90,ha="right")
fig=plt.gcf()
fig.set_size_inches(13,13)
plt.title("rating")
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x="rating",hue="type",data=veri)
plt.title("video türü ve Rating oranları")
plt.show()
```

```python
labels=["filim","TV show"]
size=veri["type"].value_counts()
colors=plt.cm.Wistia(np.linspace(0,1,2))
explode=[0,0.1]
plt.rcParams["figure.figsize"]=(9,9)
plt.pie(size,labels=labels,colors=colors,explode=explode,shadow=True,startangle=90)
plt.title("Filim  ve TV Show oranı ",fontsize=20)
plt.legend()
plt.show()
```

```python
veri["rating"].value_counts().plot.pie(autopct="%1.1f%%",shadow=True,figsize=(10,8))
plt.show()
```

```python
plt.subplots(figsize=(25,15))
wordcloud=WordCloud(
background_color="white",
    width=1920,
    height=1080
).generate(" ".join(data.country))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("country.png")
plt.show()
```
